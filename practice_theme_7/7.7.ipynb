### Игроки из каких стран играют в Российских клубах И Сколько они получают

### margins & pivot table & fill_value

```python
russian_teams_players_wages = fb.loc[
    fb['club_name'].isin(russian_clubs)
].pivot_table(
        values=['wage_eur'],aggfunc='sum',
        index=['nationality_name'], 
        columns=['club_name'], fill_value=0, #заменить пустые значения Not a Number
        margins=True)#сумма по странам и клубам (по столбцам и строкам)
russian_teams_players_wages  
```
