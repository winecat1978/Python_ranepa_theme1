Возможно - Удаление вручную с помощью маски по нужной колонке-серии и индексации по всему фрейму


```python
df.loc[:,'time'].dropna(inplace=True)

df[timeNa]
```

Удаление не распространяется на исходный фрейм при выделении нужной колонки с помощью  индексации 

```python
#dfcopy = df.loc[ : , ['id','time']].copy()
#dfcopy.dropna(axis= 0, inplace = True)
dfcopy = df.loc[ : , ['id', 'time']].copy().dropna(axis = 0 )
dfcopy
```

Удаление из фрейма с помощью параметра dropna(subset = ['colWithNa'])

```python
#df.dropna(subset= ['time', 'bet', 'win'], thresh = 3, axis = 0)
df.dropna(subset= ['time'], axis = 0, inplace=True)
df
```

```python
df.time.apply(lambda dt: dt.year)
df.time.apply(lambda dt: dt.month)
df.time.apply(lambda dt: dt.day)
df.time.apply(lambda dt: dt.strftime('%d'))
#df.time.apply(lambda dt: dt.tzinfo) # timezone
df.time.apply(lambda dt: dt.weekday()) # 0-пн
```
